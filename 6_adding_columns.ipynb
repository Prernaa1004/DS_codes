```python
import pandas as pd
import numpy as np
coffee=pd.read_csv("coffee.csv")
coffee
```

```python
#adding colums
coffee["Price"]=100
coffee
```

```python
coffee["Price"]=np.where(coffee["Coffee Type"]=="Espresso",10,100)
coffee
```

```python
coffee["Total Price"]=coffee["Units Sold"]*coffee["Price"]
coffee
```

```python
#dropping colums
#this only prints what we want it doesnt change the original data 
coffee.drop(columns=["Total Price"])
```

```python
#renaming colums
coffee.rename(columns={"Total Price":"Rev"})
coffee
```

```python
coffee=coffee.rename(columns={"Total Price":"Revenue"})
coffee
```
