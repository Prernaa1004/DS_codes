```python
#customization
data=pd.read_csv('fifa_data.csv')
barcelona=data.loc[data["Club"]=="FC Barcelona"]
madrid=data.loc[data["Club"]=="Real Madrid"]
xlabel=["FC Barcelona","Real Madrid"]

boxes=plt.boxplot([barcelona["Overall"],madrid["Overall"]],labels=xlabel,patch_artist=True)
colors=["red","green"]
facecolors=["red","green"]

for box in boxes["boxes"]:
    box.set(color=colors.pop(),linewidth=2,facecolor=facecolors.pop(0))
    

plt.show()
#or
```
