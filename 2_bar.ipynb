```python
import matplotlib.pyplot as plt
```

```python
x=["java","python","c","c++","c#"]
y=[22,17,45,37,6]

graph=plt.bar(x,y)

plt.show()
```

```python
graph=plt.bar(x,y,color='red',edgecolor='yellow',linewidth=2)
patterns=['/','o','|','x','*']
for bar in graph:
    bar.set_hatch(patterns.pop(0))

#also works as 
#graph[0].set_hatch('/')
#graph[1].set_hatch('o')
#graph[2].set_hatch('*')
plt.figure(figsize=(5,5))
plt.show()
```
