```python
import pandas as pd
bios=pd.read_csv("bios.csv")
noc=pd.read_csv("noc_regions.csv")
print(noc.head())
bios.head()
```

```python
#merging data of colums that are same i snoc file and bios file
#INNER MERGE : where both exist
pd.merge(bios,noc,left_on='born_country',right_on='NOC',how="inner").head()
```

```python
#OUTER MERGE
pd.merge(bios,noc,left_on='born_country',right_on='NOC',how="outer").head()
```

```python
#RIGHT MERGE
pd.merge(bios,noc,left_on='born_country',right_on='NOC',how="right").head()
```

```python
#LEFT MERGE 
pd.merge(bios,noc,left_on='born_country',right_on='NOC',how="left").head()
```

```python
#concat
new=[pd.concat([bios.head(),noc.head()])]
new
```
