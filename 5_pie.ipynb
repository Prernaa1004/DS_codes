```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x=["java","python","c","c++","c#"]
y=[22,17,45,37,6]
plt.pie(y,labels=x)
plt.show()
```

```python
#customization
x=["java","python","c","c++","c#"]
y=[22,17,45,37,6]
explode=[0,0,0.1,0,0]
plt.pie(y,labels=x,autopct='%1.1f%%',explode=explode,startangle=90)
#star angle is used to rotate the pie chart by 90 degree
plt.show()  

```

```python
data=pd.read_csv("data_data.csv")
left=data.loc[data["Preferred Foot"]=="Left"].count()[0]
right=data.loc[data["Preferred Foot"]=="Right"].count()[0]

labels=["Left","Right"]
values=[left,right]
colors=["red","green"]
plt.pie(values,labels=labels,autopct='%1.1f%%',colors=colors)
plt.title("Foot Preference of data Players")
plt.show()
```

```python
data=pd.read_csv("fifa_data.csv")
#print(data['Weight'].head())

#as value is in lbs that is strong format so we crreate a loop to extract only the integer part
data['Weight'] =[int(x.strip('lbs')) if type(x)==str else x for x in data.Weight]

light=data.loc[data['Weight']<125].count()[0]
light_medium=data.loc[(data['Weight']>=125) & (data['Weight']<150)].count()[0]
medium=data.loc[(data['Weight']>=150) & (data['Weight']<175)].count()[0]
medium_heavy=data.loc[(data['Weight']>=175) & (data['Weight']<200)].count()[0]
heavy=data.loc[data['Weight']>=200].count()[0]

labels=['Under 125','125-150','150-175','175-200','Over 200']
weights=[light,light_medium,medium,medium_heavy,heavy]
plt.pie(weights,labels=labels,autopct='%1.1f%%')
plt.title("Weight Distribution of data Players(in lbs)")
plt.show()
```
