```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
ages=np.random.normal(20,1.5,100)
#normal distirbution with mean 20 and standard deviation 1.5 and 100 samples
output= plt.hist(ages,bins=15)
counts=output[0]
bin_edges=output[1]

print(ages)
plt.show()
print("Counts:", counts)
print("Bin Edges:", bin_edges)
```

```python
ages=np.random.normal(20,1.5,100)
plt.hist(ages,bins=[ages.min(),16,18,21,ages.max()])
```

```python
ages=np.random.normal(20,1.5,100)
plt.hist(ages,bins=20,cumulative=True)
```

```python
#REAL LIFE DATA SET
data=pd.read_csv("fifa_data.csv")
print(data.head())
```

```python
bins=[20,30,40,50,60,70,80,90,100]

hisoutput=plt.hist(data["Overall"],bins=bins,color="red")
counts=hisoutput[0]
bins=hisoutput[1]

plt.xticks(bins)

plt.ylabel("Number of Players")
plt.xlabel("Skill Level")
plt.title("Distribution of Player Skills in FIFA 2018")

print("Counts:", counts)
print("Bin Edges:", bins)

plt.show()
```
