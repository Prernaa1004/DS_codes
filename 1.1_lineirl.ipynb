```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv('gas_prices.csv')
data.head()
```

```python
plt.plot(data["Year"],data["USA"],label="USA")
plt.plot(data["Year"],data["Canada"],label="Canada")
plt.show()
```

```python
cols=data.columns[1:]
plt.figure(figsize=(10,7))
xlabel="Year"
ylabel="Price(USD)"
plt.xlabel(xlabel)
plt.ylabel(ylabel)

for country in cols:
    plt.legend()
    plt.plot(data["Year"],data[country],label=country,marker='.',linestyle='-')
```
