```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x=[1.5,2.5,3.5]
y=[3.5,9.5,18.5]
plt.plot(x,y)

plt.title('My first plot',fontdict={'fontname':'Comic Sans MS','fontsize':20})

plt.xlabel("x label")
plt.ylabel("y label")

#i want the labels ot be intergers only and not floats
plt.xticks([1,2,3,4,10])
plt.yticks([3,7,11,15,19])



plt.show()
```

```python
#ADDING LEGENDS
x=[1.5,2.5,3.5]
y=[3.5,9.5,18.5]
plt.plot(x,y , label='Change in speed over time')

plt.title('My Second plot',fontdict={'fontname':'Comic Sans MS','fontsize':20})

plt.xlabel("Time")
plt.ylabel("Speed")
plt.legend()

plt.show()
```

```python
#adding beauty
x=[1.5,2.5,3.5]
y=[3.5,9.5,18.5]
plt.plot(x,y,color='red',linewidth=1,marker='.',markersize='20',markeredgecolor='black',markeredgewidth='2',linestyle='--')
```

```python
#shortcut for customization
#color marked linestyle
plt.plot(x,y,'go--')
```

```python
#FOR BIGGER GRAPH
plt.figure(figsize=(5,3),dpi=200)
#5 inches width and 2 inches height

#dpi is the resolution
#adding line 2
x2=np.arange(0,4.5,0.5)
print(x2)
plt.plot(x2,x2**2,'r',label='x^2')
plt.legend()

x=[0,2.5,3.5]
y=[3.5,0,18.5]
plt.plot(x,y,label='old graph')
plt.legend()


plt.savefig('mygraph.png',dpi=300)
plt.show()
```
